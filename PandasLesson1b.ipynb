```python
# 3. Sorting Data
# Ascending, Descending Order
# Numerical Columns
# Sort Data Columns -> Writing Descending Order
#
data = data.sort_values(by='Reading', ascending=False)
```

```python
# 4. Creating a New Columns: Feature Engineering
# TotolMarks = Math + English + Reading + Writing
data['TotalMarks'] = data['Reading'] + data['Writing'] + data['Math'] + data['English']
data.sort_values('TotalMarks',inplace=True, ascending = False)
data.head(10)
```

```python
# 5. Merging DataFrame
```

```python


# Assuming 'data' is your DataFrame
data['enrolldate'] = pandas.to_datetime(data['enrolldate'])
# Extracting Years and Months
data['Years'] = data['enrolldate'].dt.year
data['Months'] = data['enrolldate'].dt.month

# Printing the 'Years' and 'Months' columns
print(data['Years'])
print(data['Months'])
```

```python
data['Days'] = data['enrolldate'].dt.day
data['Days']
```

```python
# Calculate the age of students at enrollment
# Born 2000
# Enroll 2017


# Calculate the age of students at enrollment
# Assuming 'enrolldate' and 'bday' are in datetime format

# Calculate the difference between enrolldate and bday
data['AgeAtEnrollMent'] = (data['enrolldate'] - data['bday']).dt.days / 365

# Calculate the maximum and minimum age at enrollment
max_age = data['AgeAtEnrollMent'].max()
min_age = data['AgeAtEnrollMent'].min()

print(f"Maximum Age at Enrollment: {max_age} years")
print(f"Minimum Age at Enrollment: {min_age} years")
```

```python
# Generate hist of age at enrollment
import matplotlib.pyplot as plt
plt.hist(data['AgeAtEnrollMent'], bins = 20)
plt.xlabel("Age at Enrollment")
plt.ylabel("Count")
plt.title("Distribution of Age at Enrollment")
plt.show()
```
