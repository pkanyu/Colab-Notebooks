```python
# CF : Recommendations is done on attributes and characteristics of other users with similar taste.
# Trending Hashtags
# Trending News Articles
# Movie Recommendation System

# Users : How many users have viewed a particular movie
import pandas
users = pandas.read_csv("https://modcom.co.ke/datasets/file.csv")
users.head(5)
```

```python
# Movie Details
movies = pandas.read_csv("https://modcom.co.ke/datasets/Movie_Id_Titles.csv")
movies.head(5)
```

```python
# Movie Details for particular movie-id(item_id)
# Movie Title for item_id 50(n-1)
# loc


movies.loc[49]
movies.loc[171]
```

```python
# Merge users and movies in one dataframe(merge)
data = pandas.merge(users,movies,on="item_id")
data.head(5)
```

```python
# Data Analysis
# 1. Maximum Rating Per Movies
# Ans -> 5

data['rating'].max()
```

```python
# 2. Min Rating
# Ans ->1
data['rating'].min()
```

```python
# 3. Average Movie Rating
data['rating'].mean()
```

```python
# Statistics
# describe
data.describe()
```

```python
# Correlation
data.corr()
```

```python
# Modelling -> Design Recommender System
# Based:
# 1. Rating of the movies(Highest rated movie should be recommended)
# 2. Viewers(How many users viewed the movie -> Highest number of Viewers)
```

```python
# 1. Group Movies based the Average Rating
title_mean_rating = data.groupby('title')['rating'].mean()
# Which were the highest rated movies? -> Sort values in descending manner
title_mean_rating.sort_values(ascending = False).head(15)
```
