```python
# Recommender Systems:
# 1. Product Recommendation e,g Ecommerce
# 2. Media Houses: Tuko.com, recommend news articles


# Categorized into:
# a.) Content-Based Recommender - > Based on the User Past Experience(NLTK - consine similarity)
# b.) Collaborative Filtering - > Based on other review, e,g Trending Hashtag, Trending Video


# CONTENT BASED - News
# Fetch Data - News Data
import pandas
data = pandas.read_csv('https://modcom.co.ke/data/datasets/news.csv')
data.head(50)
```

```python
data.groupby("website").size()
```

```python
# Number of Records
data.shape
```

```python
# Take the first 1000 News Records for Recommendation
data = data.head(1000)
data.shape
```

```python
X = data['content'].values
```

```python
# SENTENCE TRANSFORMERS: Convert textual data
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("distilbert-base-nli-stsb-mean-tokens")
embeddings = model.encode(X, show_progress_bar=True)
```

```python
embeddings_df = pandas.DataFrame(embeddings)
embeddings_df
```
