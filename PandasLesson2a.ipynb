```python
# Types of Data (Categorical Data, Continuous)
# Data Pre Processing(Clean, Transformation, Reduction, wrangling)
```

```python
# Dropping Columns -> Dates: Time Series Analysis
data.dropna(subset=["bday","enrolldate","expgradate"], inplace = True)
```

```python
data.isnull().sum()
```

```python
# Filling Empties for Categorical Data:
# Gender
# Smoking
# LiveOnCampus

# 1. Gender
# Empties -> Unknown
data.groupby("Gender").size()
data['Gender'].fillna('Unkown',inplace = True)
data.groupby('Gender').size()
```

```python
# Smoking
data.groupby('Smoking').size()
data['Smoking'].fillna('Unknown',inplace = True)
data.groupby('Gender').size()
```

```python
# EDA: Explanatory Data Analysis and Time Series Analysis(Date and Time)
# selecting columns:
# Single: Syntax -> data['column']
# Multiple Column -> data[['column1',column2]]
data[['Gender','Major']]
```

```python
#2. Filtering
# Filtering the DataFrame Based on a Condition
# Students who score more than 90 in maths
# Syntax: dataFrame[condition]

scored_90_maths = data[data['Math'] > 90]
scored_90_maths
```

```python
# More Filtering:
# Freshman and Scored More than 86 in English
data[(data['English'] >90) & (data['Rank'] == 'Freshman')]
```

```python
# Handling Times and Dates: Time Series Analysis
data['enrolldata'] = pandas.to_datetime(data['enrolldate'])
# Extracting Years
data['Years'] = data['enrolldate'].dt.year
data['month'] = data['enrolldate'] .dt.month
data['Days'] = data['enrolldate'].dt.days

data['Years']
data['Months'].head(50)
```

```python
# Calculate the age of student at enrollment
# Born 2000
# Enroll 2017
data['AgeAtEnrollMent'] = (data['enrolldate'] - data['bday'])dt.days/365
```

```python
# The Number of Enrollment Each Year

enrollments_per_year = data['enrolldata'].dt.year.value_count().sort_index()
enrollments_per_year
```
