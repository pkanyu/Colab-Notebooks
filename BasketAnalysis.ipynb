```python
# Basket Analysis
# Machine Learning
# Supervised Learning -> Regression, Classification
# Unsupervised Learning -> Clustering, Association Rules/Data Mining(Basket Analysis)
# Basket Analysis: A technique that applys association rules(apriori) to discover relationship between item
```

```python
# Data Collection
import pandas
data = pandas.read_csv("https://modcom.co.ke/data/datasets/store.csv")
data
```

```python
# Import association algorithms(apriori algorithms)
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Data Analysis
# 1. Transactions Per Country
data.groupby(
  'Country'
).size()
```

```python
# Market Analysis: Purchasing taste depends on location
# Urban Setting: What is purchased include e,g Pizza, Cannded Food...
# Rural Setting: Vegetables....
```

```python
# Clothes that are bought in the US(polo t-shirts) are very different from clothes bought in Arabian Countries(kanzus)
# Adjust a particular location (UK)
data = data[data['Country'] == 'France']
data
```

```python
# Null Records
# InoviceNo, Description, Quantity, CustomerID
# 66 customers in France Transaction Paid By Cash(ID Not Recorded)
data.isnull().sum()
```

```python
# Groupby - InovoiceNo, Description alongside the Total Quantity
# A Negative Quantity Represents the Returns
#
data.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
```

```python
# Create a pivot table and Dump all the Transaction(Single Pool)
# index -> InvoiceNo
# columns -> Description
# values -> Quantity
# Replace Quantity NaN with 0
pandas.set_option('display.max_columns',300)
baskets = data.pivot_table(index = 'InvoiceNo', columns='Description', values = 'Quantity',fill_value=0)
baskets
```

```python
# Any Item that has a quantity more than 1: We will mark that it was bought(label it to 1)
# 0 represent: Item Not Bought
# 1 represent: Item Bought

# Create a function(if quantity > 1 mark1 else mark 0)
def encode(x):
  if x <= 0:
    return 0
  if x > 0:
    return 1
```

```python
# Apply to the Dataset Pivot
basket_set = baskets.applymap(encode)
basket_set
```

```python
# Apply the Association Rules(apriori)
# Frequent Item Set
# support 0.07 -> 14% of the total transactions
frequent_items_sets = apriori(basket_set, min_support = 0.07,use_colnames=True)
rules = association_rules(frequent_items_sets,metric = 'lift')
rules[(rules['lift'] > 7) & (rules['confidence'] > 0.7) ]
# Support
# Confidence
# Lift
```
