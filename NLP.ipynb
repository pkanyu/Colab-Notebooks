```python
# Natural Language Processing: Branch in AI that involves interacting with machines(phones, car, laptop, fridge, robots) using
# a Natural Language(text, speech) rather than a code.
# Application Areas: Robotics, ChatBots(zuri,apple siri, amazon alexa)
# Data Science vs NLP: Analyse text, speech(draw conclusion) e,g Sentiment Analysis, Web Scrapping

# Text Analysis and WordCloud
# Web Scrapping
# Twitter Sentiment Analysis
```

```python
# Get text data
text = ''' For weeks the European Union's stance on the war has been clouded with mixed messages, diplomatic gaffes and conflicting national views on the Israeli-Palestinian conflict.'''
```

```python
# Step1: import nlkt(Natural Language Processing Tool) ->python library
import nltk
nltk.download('punkt')
```

```python
# Convert text to sentences or words, or phrases -> Token
# Tokenization
from nltk.tokenize import sent_tokenize, word_tokenize
tokenized_text = word_tokenize(text)
print(tokenized_text)
```

```python
# Removing Stop Words:
# Stops Words: Connecting words in sentences that dont carry meaning context
nltk.download('stopwords')
```

```python
# Generate the fields -> Different Languages
from nltk.corpus import stopwords
print(stopwords.fileids())

english_stopwords = set(stopwords.words('english'))
print(english_stopwords)
```

```python
from numpy import mean
meanful_words = []
for word in tokenized_text:
  if word.lower() not in english_stopwords:
    meanful_words.append(word)
# meanful_word: contains all words without stopwords
print(meanful_words)
```

```python
# Clean the Words: By Removing all the punctuations e,g ,..!...
clean_words = []
for word in meanful_words:
  if word.isalnum():
    clean_words.append(word)


print(clean_words)
```

```python
# Frequency Distribution: Statistical representation of how often different values or elements occur in a dataset.
# Goal: To summarize and understand the distribution of data.
from nltk.probability import FreqDist
frequency_count = FreqDist(clean_words)
frequency_count.most_common(20)
```

```python
# Visualize the FreqDist on Wordcloud
from wordcloud import WordCloud, STOPWORDS
from PIL import Image

# create a function to generate WordCloud
def generate_wordcloud(text):
  stopwords = set(STOPWORDS)
  wc = WordCloud(background_color="white",max_words=1000,stopwords=stopwords,repeat=True)
  wc.generate(str(text))
  wc.to_file("wc.png")
  path = "wc.png"
  display(Image.open(path))
```

```python
generate_wordcloud(frequency_count.most_common(20))
```
